## Connecting to the Database

```python
# Your code here
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite')
cur = conn.cursor()
```

## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here
cur.execute("SELECT * FROM employees JOIN offices USING(officeCode) WHERE city = 'Boston'")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Your code here
cur.execute("SELECT officeCode, city, COUNT(employeeNumber) as n_employees FROM employees JOIN offices USING(officeCode) GROUP BY officeCode;")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
cur.execute("""SELECT productName
               FROM products 
               JOIN orderdetails USING(productCode)
               JOIN orders USING(orderNumber)
               JOIN customers c USING(customerNumber)
               JOIN employees e 
               ON c.salesRepEmployeeNumber = e.employeeNumber;
               """)
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head()
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
cur.execute("""SELECT firstName, lastName, COUNT(orderNumber) as Sales
               FROM products 
               JOIN orderdetails USING(productCode)
               JOIN orders USING(orderNumber)
               JOIN customers c USING(customerNumber)
               JOIN employees e 
               ON c.salesRepEmployeeNumber = e.employeeNumber
               GROUP BY lastName
               ORDER BY Sales DESC;
               """)
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```
